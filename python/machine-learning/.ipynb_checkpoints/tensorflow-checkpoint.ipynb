```python
def sigmoid(x):
    return 1/(1+ np.exp(-x))

def relu(x):
    return np.maximum(x, 0)

x = np.arange(-5., 5., 0.2)
plt.subplot(121)
plt.title('sigmoid')
plt.plot(x, sigmoid(x))

plt.subplot(122)
plt.title('relu')
plt.plot(x, relu(x))


```
